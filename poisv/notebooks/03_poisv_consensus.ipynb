# PoISV Consensus Demo


```python
from agentsprotocol.s_con import compute_s_con
from agentsprotocol.psi_test import compute_psi, check_acceptance
from agentsprotocol.validator import block_weight
s = [compute_s_con(t, ['Paris is the capital of France'], tau=0.7) for t in ['Paris is the capital of France', 'The sky is blue']]
ev = [[0.1, 0.2, 0.3, 0.1], [0.3, 0.1, 0.1, 0.2], [0.2, 0.3, 0.15, 0.25]]
psi = compute_psi(ev)
print('accepted =', check_acceptance(s, psi, theta_min=0.3, psi_min=0.3))
print('weight   =', block_weight(psi, s))
```
